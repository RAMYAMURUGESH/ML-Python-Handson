### Melt

```python
df = pd.read_csv('life_expectancy_years.csv')
```

```python
df.head()
df.shape #its in wide format 187 rows nd 302 columns
```

```python
df.columns
```

```python
melt_df = df.melt(id_vars='country',var_name='year',value_name='life_expectancy')
melt_df
```

## pivot 

```python
pivot_df = melt_df.pivot(index='country',columns=['year'],values = 'life_expectancy')
pivot_df
```

```python
ab=pd.read_csv('employee.csv')
ab.head()
ab['DistanceFromHome'].value_counts()

```

```python
def modify(x):
    if x< 2:
        return 'near'
    elif x>2 and x<5:
        return 'far'
    else: 
        return 'very far'
    
```

```python
ab['DistanceFromHome']= ab['DistanceFromHome'].apply(modify)
ab['DistanceFromHome'].value_counts()
```

```python
ab['Age'].value_counts()
```

```python
def age_modify(x):
    if x <20:
        return 'teen'
    elif x>20 and x<40:
        return 'adult'
    elif x>40:
        return 'old age'
```

```python
ab['Age_grp'] = ab['Age'].apply(age_modify)
ab['Age_grp'].value_counts()
```
